```python
rotate_lists=glob.glob('/home/zenalisa/data/50ar_last/lists/*rotate.csv')
print(len(rotate_lists))
for r in rotate_lists:
    rlines=open(r, 'r').readlines()[1:]
    outcsv=open(r.replace('lists', ''), 'w')
    outcsv.write('path_signal, path_target\n')
    for l in tqdm(rlines):
        aia_new=l.split(',')[0].strip().replace('/50ar_last', '/50ar_last/rotate')
        hmi_new=l.split(',')[1].strip().replace('/50ar_last', '/50ar_last/rotate')
        aia_new_save=aia_new.replace('/fs1/project/iip', '/home')
        hmi_new_save=hmi_new.replace('/fs1/project/iip', '/home')
        
        rotate=np.random.randint(0, 2)
        #print(rotate)
        rotate=bool(rotate)
        #print(rotate)
        
        aia_file=l.split(',')[0].strip().replace('/fs1/project/iip', '/home')
        hmi_file=l.split(',')[1].strip().replace('/fs1/project/iip', '/home')
        a=np.array(Image.open(aia_file))
        h=np.array(Image.open(hmi_file))
        
        if rotate:
            a=np.rot90(a)
            a=np.rot90(a)
            h=np.rot90(h)
            h=np.rot90(h)
            
        
        #out_a = np.array(a, dtype=np.uint8)
        #out_h = np.array(h, dtype=np.uint8)
        outa=Image.fromarray(a)
        outh=Image.fromarray(h)
        #print(aia_new_save.replace(aia_new_save.split('/')[-1], ''))
        #raise KeyError
        if not os.path.exists(aia_new_save.replace(aia_new_save.split('/')[-1], '')):
            os.makedirs(aia_new_save.replace(aia_new_save.split('/')[-1], ''))
        
        if not os.path.exists(hmi_new_save.replace(hmi_new_save.split('/')[-1], '')):
            os.makedirs(hmi_new_save.replace(hmi_new_save.split('/')[-1], ''))
        
        outa.save(aia_new_save)
        outh.save(hmi_new_save)
    
        outcsv.write(aia_new+ ','+ hmi_new+'\n')
    outcsv.close()
    #raise KeyError
```
